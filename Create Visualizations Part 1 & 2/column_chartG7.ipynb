**Column Chart**

```python
import matplotlib.pyplot as plt

medalcounts_gender = df['Gender'].value_counts()

plt.figure(figsize=(6, 6))
medalcounts_gender.plot(kind ='bar')
plt.title('Number of Medals Won by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Medals')

plt.show()

#column chart is vertical & bar chart is horizontal
#number of medals won by women and men
```

**Line plot**

```python
country_data = df[df['Country'] == 'Italy']


medal_counts_by_year = country_data.groupby('Year').size().reset_index(name='Medal Count')


plt.figure(figsize=(10, 6))
plt.plot(medal_counts_by_year['Year'], medal_counts_by_year['Medal Count'])

plt.title('Number of Medals Won by Italy Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Medals')



plt.show()
#line plot- It shows trend over the period of time
#trend in the number of medals won by Italy over the years
#spike in 2005
```

**Scatter Plot**

```python
country_name = 'United States'


country_data = df[df['Country'] == country_name]


medals_by_year = country_data.groupby('Year').size().reset_index(name='Medal Count')


plt.figure(figsize=(12, 6))
plt.scatter(medals_by_year['Year'], medals_by_year['Medal Count'], alpha=0.7, c='skyblue', edgecolor='black')
plt.title(f'Medals Won by United States Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.xticks(rotation=45)
plt.grid(True)


plt.show()
#scatter plot
#two numeric values
```

**Waffle Chart**

```python
from pywaffle import Waffle

medal_counts = df['Medal'].value_counts()


fig = plt.figure(
    FigureClass=Waffle,
    rows=5,
    columns=11,
    values=medal_counts,
    colors =["#CD7F32","#FFD700", "#C0C0C0"],
    title={'label': 'Distribution of Medals by Type', 'loc': 'center'},
    labels=["{} ({})".format(k, v) for k, v in medal_counts.items()],
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)}
)
plt.show()

#waffle chart
#distribution of medal types, showing how they contribute to the total
```

**Chloropeth Map**

```python
import plotly.express as px


medals_by_country = df['Country'].value_counts().reset_index()
medals_by_country.columns = ['Country', 'Medals']

fig = px.choropleth(
    medals_by_country,
    locations="Country",
    locationmode="country names",
    color="Medals",
    hover_name="Country",
    color_continuous_scale=px.colors.sequential.Plasma,
    title="Number of Medals Won by Country"
)
fig.show()

#cholropeth map
#A map of the world with countries colored by the number of medals they have won
```

**Pie Chart**

```python
gender_counts = df['Gender'].value_counts()


plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])


plt.title('Gender Distribution of Athletes')


plt.show()

#pie chart
#proportion of male and female athletes in the dataset
```

**Treemap**

```python
treemap_data = df[['Country', 'Sport', 'Medal']]
medals_by_country_sport = treemap_data.groupby(['Country', 'Sport']).size().reset_index(name='Medals')

#treemap
fig = px.treemap(
    medals_by_country_sport,
    path=['Country', 'Sport'],
    values='Medals',
    color='Medals',
    color_continuous_scale='Viridis',
    title='Distribution of Medals by Country and Sport'
)
fig.show()

#treemap
```

**Area Chart**

```python
medals_by_year = df.groupby('Year').size()

#area chart
plt.figure(figsize=(10, 6))
plt.fill_between(medals_by_year.index, medals_by_year.values, color='skyblue', alpha=0.4)
plt.plot(medals_by_year.index, medals_by_year.values, color='Slateblue', alpha=0.6, linewidth=2)
plt.title('Number of Medals Won by Year')
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.grid(True)
plt.show()

#area chart
#shows how the number of medals has increased
```

**Word Cloud**
